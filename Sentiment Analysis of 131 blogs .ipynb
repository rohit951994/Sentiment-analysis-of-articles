```python
def extract_title_and_article(number, url):
    try:
        source = requests.get(url)
        soup = BeautifulSoup(source.text, 'html.parser')

        title_element = soup.find('h1', class_='entry-title')
        title = title_element.text 

        articles_element = soup.find('div', class_='td-post-content tagdiv-type')
        articles = articles_element.text 

    except:
        source = requests.get(url)
        soup = BeautifulSoup(source.text, 'html.parser')

        title_element = soup.find('h1', class_='tdb-title-text')
        title = title_element.text if title_element else "page error"

        articles_element = soup.find('div', class_='td_block_wrap tdb_single_content tdi_130 td-pb-border-top td_block_template_1 td-post-content tagdiv-type')
        articles = articles_element.text if articles_element else "page error"
    
    path = r'New folder/text files'
    file_name = f"{number}.txt"
    with open(os.path.join(path,file_name), 'w', encoding='utf-8') as file:
        file.write(f"{title}\n")
        file.write(f"{articles}\n")

    print(f"Title and article extracted from {url} and written to {file_name}")
```

```python
for  _,row in input_df.iterrows():
    number = row['URL_ID']
    url = row['URL']
    extract_title_and_article(number, url)
```

# Parameters about the textual data

```python
# replace new line with fullstop.
def cleaning_content(x):
    x = x.replace('\n',' ')
    x=x.replace('?','.')
    x = re.sub('[^a-zA-Z0-9 \.]', '', x)
    x = re.sub('\s{2,}', " ", x)
    x=x.lower()
    return x
    
```

```python
df['Content']=df['Content'].apply(cleaning_content)
```

```python
def avg_sentence_length(x):
    sentences = re.split(r'(?<=[.])\s+', x)
    num_sentences = len(sentences)

    # Count the number of words
    words = x.split()
    num_words = len(words)

    # Calculate average sentence length
    average_sentence_length = num_words / num_sentences
    return average_sentence_length
```

```python
df['AVG SENTENCE LENGTH']=df['Content'].apply(avg_sentence_length).round(0)
```

```python
df['AVG NUMBER OF WORDS PER SENTENCE']=df['AVG SENTENCE LENGTH']
```

```python
def avg_word_length(x):
    words = list(x.split())
    no_of_words=len(words)
    word_length = [len(word) for word in words]
    avg_word_length=sum(word_length)/no_of_words
    return avg_word_length
```

```python
df['AVG WORD LENGTH']=df['Content'].apply(avg_word_length).round(2)
```

```python
def count_syllables(word):
    # Handling exceptions for words ending with "es" or "ed"
    if word.endswith("es") or word.endswith("ed"):
        vowels = re.findall(r'[aeiou]', word)
        vowel_count=len(vowels)-1
        return vowel_count

    # Counting the number of vowel sequences
    vowels = re.findall(r'[aeiou]', word)
    vowel_count=len(vowels)
    return vowel_count
```

```python
def syllables_per_word(x):
    word_count=len(x.split())
    words = re.findall(r'\b\w+\b', x)
    syllables_per_word = [count_syllables(word) for word in words]
    avg_syllable_per_word=sum(syllables_per_word)/word_count
    return avg_syllable_per_word
```

```python
def complex_word_count(x):
    count=0
    words = re.findall(r'\b\w+\b', x)
    syllables_per_word = [count_syllables(word) for word in words]
    
    for syllable in syllables_per_word:
        if syllable>2:
            count+=1
    return count       
```

```python
def percentage_complex_word(x):
    count=0
    word_count=len(x.split())
    words = re.findall(r'\b\w+\b', x)
    syllables_per_word = [count_syllables(word) for word in words]
    
    for syllable in syllables_per_word:
        if syllable>2:
            count+=1
    percentage_complex_word=round (count/word_count,2)
    return percentage_complex_word
```

```python
df['SYLLABLE PER WORD']=df['Content'].apply(syllables_per_word).round(2)
```

```python
df['COMPLEX WORD COUNT']=df['Content'].apply(complex_word_count).round(2)
```

```python
df['PERCENTAGE OF COMPLEX WORDS']=df['Content'].apply(percentage_complex_word).round(2)
```

```python
df['FOG INDEX']=(df['AVG SENTENCE LENGTH']+df['PERCENTAGE OF COMPLEX WORDS'])*0.4
```

```python
# lemmatization of content
def lemmatize_text(x):
    x = re.sub(r'[^a-zA-Z\s]', ' ', x)
    x=x.lower()
    x=x.replace('\n',' ')
    x = re.sub('\s{2,}', " ", x)
    tokens = word_tokenize(x)
    words = [t for t in tokens if t not in stop_words]
    lemmatize = [lemma.lemmatize(w) for w in words]
    return lemmatize
```

```python
df_senti['processed_content']=df_senti['Content'].apply(lemmatize_text)
```

```python
df_senti.drop(columns='Content',inplace=True)
df_senti
```

```python
df_senti['POSITIVE SCORE'] = df_senti['processed_content'].map(lambda x: len([i for i in x if i in positive_words_final]))

df_senti['NEGATIVE SCORE'] = df_senti['processed_content'].map(lambda x: len([i for i in x if i in negative_words_final]))
```

```python
df_senti
```

```python
df_senti['WORD COUNT'] = df_senti['processed_content'].map(lambda x: len(x))
```

```python
df_senti['POLARITY SCORE']=(df_senti['POSITIVE SCORE']-df_senti['NEGATIVE SCORE'])/((df_senti['POSITIVE SCORE']+df_senti['NEGATIVE SCORE'])+.000001)
df_senti['SUBJECTIVITY SCORE']=(df_senti['POSITIVE SCORE']+df_senti['NEGATIVE SCORE'])/(df_senti['WORD COUNT']+.000001)
```
